#### a) Outlier detection

```python
Q1 = df_phones.select_dtypes('number').quantile(0.25)
Q3 = df_phones.select_dtypes('number').quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
potential_outliers = (df_phones.select_dtypes('number') < lower_bound) | (df_phones.select_dtypes('number') > upper_bound)
potential_outliers.sum()
```

#### significant number of outliers seem to be present in the below columns
        1) price
        2) battery capacity
        3) internal memory
        4) screen size
        5) primary camera front
        6) resolution_width

### 3) Data Preprocessing

```python
df_m = df_phones.copy()
```

##### a) columns to drop

```python
df_m.columns
```

```python
df_m = df_m.drop(['fast_charging','model','brand_name'],axis=1)
```

```python
df_m.head()
```

```python
df_m.shape
```

##### b) drop unencessary rows

```python
df_m = df_m.dropna()
```

```python
df_m = df_m.drop_duplicates()
```

```python
df_m.shape
```

##### c) drop outliers

```python
outlier_cols = ['battery_capacity','internal_memory','screen_size','primary_camera_rear','primary_camera_front','resolution_height','resolution_width']
```

```python
Q1 = df_m[outlier_cols].quantile(0.25)
Q3 = df_m[outlier_cols].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound = Q3+1.5*IQR
print(lower_bound)
print(upper_bound)
```

```python
df_m = df_m[~((df_m['battery_capacity']<lower_bound['battery_capacity'])|(df_m['battery_capacity']>upper_bound['battery_capacity']))]
df_m = df_m[~((df_m['internal_memory']<lower_bound['internal_memory'])|(df_m['internal_memory']>upper_bound['internal_memory']))]
df_m = df_m[~((df_m['screen_size']<lower_bound['screen_size'])|(df_m['screen_size']>upper_bound['screen_size']))]
df_m = df_m[~((df_m['primary_camera_rear']<lower_bound['primary_camera_rear'])|(df_m['primary_camera_rear']>upper_bound['primary_camera_rear']))]
df_m = df_m[~((df_m['primary_camera_front']<lower_bound['primary_camera_front'])|(df_m['primary_camera_front']>upper_bound['primary_camera_front']))]
df_m = df_m[~((df_m['resolution_height']<lower_bound['resolution_height'])|(df_m['resolution_height']>upper_bound['resolution_height']))]
df_m = df_m[~((df_m['resolution_width']<lower_bound['resolution_width'])|(df_m['resolution_width']>upper_bound['resolution_width']))]
```
